```python
# seems like there is only one player data is duplicate 

Fifa_df = Fifa_df.drop_duplicates()
```

```python
Fifa_df[Fifa_df.duplicated()]
```

```python
# Rename some column names

Fifa_df.rename(columns={
    '↓OVA': 'Rating ★',
    'playerUrl': 'PlayerUrl',
    'photoUrl': 'PhotoUrl',
    'Value': 'Market Value €',
    'Weight': 'Weight in Kgs',
    'Wage': 'Weekly Wages €',
    'foot': 'Foot',
    'Joined': 'Date of Join',
    'BP': 'Best Position'
}, inplace=True)

Fifa_df.head()
```

```python
# remove the /n in team & contract column 

Fifa_df['Team & Contract'] = Fifa_df['Team & Contract'].str.replace('\n', ' ')
Fifa_df['Team & Contract'] = Fifa_df['Team & Contract'].str.strip()
Fifa_df.head()
```

```python
# Separate the Team & Contract Column into Team name and Contract year 

Fifa_df[['Club Name','Contract Duration']] = Fifa_df['Team & Contract'].str.extract(r'(.+)\s(\d{4} ~ \d{4})')
Fifa_df[['Club Name','Contract Duration']].head()
```

```python
# converting Date of join to Date format and change the format 

Fifa_df['Date of Join'] = pd.to_datetime(Fifa_df['Date of Join'], format='%b %d, %Y')
Fifa_df['Date of Join'] = Fifa_df['Date of Join'].dt.strftime('%d/%m/%Y')
```

```python
Fifa_df['Date of Join'][:5]
```

```python
# convert the rating out of 100 into star rating out of 5

Fifa_df['Rating ★'] = (Fifa_df['Rating ★'] / 20).round(1)
Fifa_df['Rating ★'].head()
```

```python
# Convert the Weight lbs into Kgs 

Fifa_df['Weight in Kgs'] = Fifa_df['Weight in Kgs'].apply(lambda x: int(x.split('lbs')[0]) * 0.453592).round().astype(int)
Fifa_df['Weight in Kgs'][:5]
```

```python
# Remove the € in the market value column

Fifa_df['Market Value €'] = Fifa_df['Market Value €'].str.replace('€', '')
Fifa_df['Market Value €'][:6]
```

```python
# Remove the € in the Weekly Wages column

Fifa_df['Weekly Wages €'] = Fifa_df['Weekly Wages €'].str.replace('€', '')
Fifa_df['Weekly Wages €'][:6]
```

```python
# Replace the spaces between the Positions into comma 
Fifa_df['Positions'] = Fifa_df['Positions'].str.replace(' ', ', ')
```

```python
# Drop the Team & Contract Column

Fifa_df.drop(columns=['Team & Contract'], inplace=True)
```

```python
columns = ['ID','PhotoUrl','PlayerUrl','LongName','Name','Nationality',
            'Age','Height','Weight in Kgs','Club Name',
           'Contract Duration','Date of Join','Positions','Best Position','Foot'
          ,'Market Value €','Weekly Wages €', 'Rating ★'] 
```

```python
Fifa_df = Fifa_df[columns]
```
